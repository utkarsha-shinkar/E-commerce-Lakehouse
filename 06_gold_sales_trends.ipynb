```python
# Daily Sales Trend
from pyspark.sql.functions import sum, count, countDistinct, col, date_format

gold_daily_sales_df = (
    spark.table("main.ecommerce.fact_sales")
    .groupBy("order_date")
    .agg(
        sum("revenue").alias("daily_revenue"),
        countDistinct("order_id").alias("daily_orders"),
        count("*").alias("daily_items")
    )
    .orderBy("order_date")
)

#display(gold_daily_sales_df)

# Write to Delta table
gold_daily_delta_table = gold_daily_sales_df.write\
    .format("delta")\
    .mode("overwrite")\
    .saveAsTable("main.ecommerce.gold_daily_sales")
```

```python
# Monthly Sales Trend
from pyspark.sql.functions import sum, count, countDistinct, col, date_format

gold_monthly_sales_df = (
    spark.table("main.ecommerce.fact_sales")
    .withColumn("year_month", date_format("order_date", "yyyy-MM"))
    .groupBy("year_month")
    .agg(
        sum("revenue").alias("monthly_revenue"),
        countDistinct("order_id").alias("monthly_orders")
    )
    .orderBy("year_month")
)

#display(gold_monthly_sales_df)

# Write to Delta table
gold_monthly_delta_table = gold_monthly_sales_df.write\
    .format("delta")\
    .mode("overwrite")\
    .saveAsTable("main.ecommerce.gold_monthly_sales")
```
