```python
# View Delta Table History (Audit Log)

fact_history = spark.sql("DESCRIBE HISTORY main.ecommerce.fact_sales")

display(fact_history)
```

```python
# Query Previous Versions (Time Travel)

fact_version =spark.read \
    .format("delta") \
    .option("versionAsOf", 0) \
    .table("main.ecommerce.fact_sales")

display(fact_version)
```

```python
%sql
--Simulate an Incorrect Update (Intentional Mistake)

-- STEP 1 ---

-- 1. Wrong update
UPDATE main.ecommerce.fact_sales
SET revenue = 0;

-- 2. Validate the mistake
SELECT SUM(revenue) FROM main.ecommerce.fact_sales;
```

```python
%sql

-- Step 2 :Restore Table Using VERSION

-- 1. Find correct version
DESCRIBE HISTORY main.ecommerce.fact_sales;

-- 2: Restore
RESTORE TABLE main.ecommerce.fact_sales TO VERSION AS OF 1;
```

```python
%sql

-- Step 3:Validate Successful Rollback

-- 1: Revenue check
SELECT SUM(revenue) FROM main.ecommerce.fact_sales;

-- 2: Check history again
DESCRIBE HISTORY main.ecommerce.fact_sales;
```
