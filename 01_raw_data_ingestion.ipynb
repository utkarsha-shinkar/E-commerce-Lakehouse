```python
customer_df = spark.read\
    .option("header",True)\
    .csv("/Volumes/main/ecommerce/lakehouse_vol/Raw_DS/olist_customers_dataset.csv")
#display(customer_df)

items_df = spark.read\
    .option("header",True)\
    .csv("/Volumes/main/ecommerce/lakehouse_vol/Raw_DS/olist_order_items_dataset.csv")
#display(items_df)

payment_df = spark.read\
    .option("header",True)\
    .csv("/Volumes/main/ecommerce/lakehouse_vol/Raw_DS/olist_order_payments_dataset.csv")
#display(payment_df)

orders_df = spark.read\
    .option("header",True)\
    .csv("/Volumes/main/ecommerce/lakehouse_vol/Raw_DS/olist_orders_dataset.csv")
#display(orders_df)

product_df = spark.read\
    .option("header",True)\
    .csv("/Volumes/main/ecommerce/lakehouse_vol/Raw_DS/olist_products_dataset.csv")
#display(product_df)
```

```python
customer = customer_df.write\
    .mode("overwrite")\
    .format("parquet")\
    .save("Volumes/main/ecommerce/lakehouse_vol/bronze/customers/")

items = items_df.write\
    .mode("overwrite")\
    .format("parquet")\
    .save("Volumes/main/ecommerce/lakehouse_vol/bronze/order_items/")

payments =payment_df.write\
    .mode("overwrite")\
    .format("parquet")\
    .save("Volumes/main/ecommerce/lakehouse_vol/bronze/payments/")


orders = orders_df.write\
    .mode("overwrite")\
    .format("parquet")\
    .save("Volumes/main/ecommerce/lakehouse_vol/bronze/orders/")

products = product_df.write\
    .mode("overwrite")\
    .format("parquet")\
    .save("Volumes/main/ecommerce/lakehouse_vol/bronze/products/")
```
