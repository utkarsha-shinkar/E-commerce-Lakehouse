```python
# Top Products by Revenue
from pyspark.sql.functions import sum, count, countDistinct, col, date_format

gold_top_products_df = (
    spark.table("main.ecommerce.fact_sales")
    .groupBy("product_id", "product_category_name")
    .agg(
        sum("revenue").alias("total_revenue"),
        countDistinct("order_id").alias("total_orders"),
        count("*").alias("total_items")
    )
    .orderBy(col("total_revenue").desc())
)

#display(gold_top_products_df)

# Write to Delta table
gold_top_delta_table = gold_top_products_df.write\
    .format("delta")\
    .mode("overwrite")\
    .saveAsTable("main.ecommerce.gold_top_products")
```
